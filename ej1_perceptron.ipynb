```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("TP3-1.csv")
df
```

```python
data_ls = df.to_numpy()
data_ls
```

```python
class_A = [(x, y) for x, y, label in data_ls if label == 1]
class_B = [(x, y) for x, y, label in data_ls if label == -1]
display(class_A, class_B)
```

```python
class_A_x, class_A_y = zip(*class_A)
class_B_x, class_B_y = zip(*class_B)
```

```python
#sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))
plt.scatter(class_A_x, class_A_y, c='b', marker='o', label='Clase A')
plt.scatter(class_B_x, class_B_y, c='r', marker='x', label='Clase B')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.title('Datos Linealmente Separables')
plt.show()
```
